```python
ch= ['Name', 'Score', 'Status']
```

```python
def status(score,lb,ub):
  if score>=lb and score<=ub:
    return 'selected'
  else:
    return 'not selected'
df['Status']=sdmerge['sp'].apply(lambda x:status(x,lb,ub))
df['Name']=sdmerge['sname']
df['Score']=sdmerge['sp']
df
```

```python
import pandas as pd
import numpy as np
```

```python
afm= pd.read_csv("/content/ahamovies.csv")
```

```python
afm.head()
```

```python
afc= pd.read_csv("/content/ahacredits.csv")
```

```python
afc.head()
```

```python
afm=afm.rename(index=str, columns={"movieid": "id"})
```

```python
afm.head()
```

```python
print("Movies:",afm.shape)
print("Survey:",afc.shape)
```

```python
afmg=afm.merge(afc, on='id')
afmg.head()
```

```python
afmg.dropna(axis=1,inplace=True)
afmg
```

```python
afmg.info()
```

```python
p=afmg['votecount']
q=afmg['voteavg']
r=afmg['voteavg'].mean()
s=afmg['votecount'].quantile(0.70)
```

```python
afmg['average']=((q*p)+(r*s))/(p+s)
```

```python
afmg.head()
```

```python
movierank=afmg.sort_values('average',ascending=False)
```

```python
movierank[['movietitle', 'votecount', 'voteavg','average','cast']].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
avg=movierank.sort_values('average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=avg['average'].head(5), y=avg['movietitle'].head(5), data=avg,palette='bright')
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
avg=movierank.sort_values('average',ascending=False)
plt.figure(figsize=(10,4))
axis1=sns.barplot(x=avg['average'].head(5), y=avg['movietitle'].head(5), data=avg,hue=avg['cast'].head(5))
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
avg=movierank.sort_values('average',ascending=True)
plt.figure(figsize=(8,4))
axis1=sns.barplot(x=avg['average'].head(3), y=avg['movietitle'].head(3), data=avg,hue=avg['cast'].head(3))
plt.xlim(4, 10)
plt.title('Underrated Movies by votes', weight='bold')
plt.xlabel('Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('underratedmovies.png')
```
